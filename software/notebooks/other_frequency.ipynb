```python
import os
import matplotlib.pyplot as plt
import numpy as np
import math
import random
import csv
top_level=f'{os.getcwd()}/..'
```

# make 2GHz modulation

```python
top_level=f'{os.getcwd()}/..'
path = f'{top_level}/data/20GspsOscope2GHzSignal.csv'

data_20gsps = []

with open(path) as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quoting=csv.QUOTE_NONE)
    for i, row in enumerate(reader):
        data_20gsps.append(float(row[1]))

t=np.arange(0,len(data_20gsps))*1000/20000
plt.plot(t,data_20gsps)
plt.xlim(0,10)
plt.show()
```

```python
timestep=t
newTimeBin = 1.0/(0.509*6.0) # Unit of ns for 3.054 GSPS
newSize    = 32000
newTimeSteps = np.linspace(0, newTimeBin*(newSize-1), num=newSize)
newwaveform = np.interp(newTimeSteps, timestep, data_20gsps)
plt.figure(figsize=(20,5))
plt.plot(t[:32000],newwaveform)
plt.xlim(0,20)
plt.grid()
plt.show()
```

```python
#FFT
size    = len(newwaveform) # Number of data points in the file
timeBin = 1000/4072 # Unit of ns
freqBin = (0.5E+3/timeBin)/float(size//2) # Units of MHz
freqSteps = np.linspace(0, freqBin*(size//2-1), num=size//2)
freq = np.fft.fft(newwaveform)/float(size)
freq = freq[range(size//2)]
mag = 20.0*np.log10(np.abs(freq)/32767.0) # Units of dBFS

plt.plot(freqSteps[0:20000],mag[0:20000])
plt.xlabel("Frequency(MHz)")
plt.grid()
```

#  make 1GHz modulation

```python
a=np.array([0,0.05,0.3,0.4,0.5,0.6,0.8,1,1,0.8,0.6,0.5,0.4,0.3,0.05,0])
t=np.arange(0,16*100)*1000/4072
for i in range(100):
    waveform=np.tile(a,100)
    waveform=waveform*np.sin(2*np.pi*1*t)

plt.figure(figsize=(10,5))
plt.plot(t,waveform)
plt.xlim(0,20)
plt.xlabel('time (ns)')
plt.grid()
plt.show()
```

```python
#FFT
size    = len(waveform) # Number of data points in the file
timeBin = 1000/4072 # Unit of ns
freqBin = (0.5E+3/timeBin)/float(size//2) # Units of MHz
freqSteps = np.linspace(0, freqBin*(size//2-1), num=size//2)
freq = np.fft.fft(waveform)/float(size)
freq = freq[range(size//2)]
mag = 20.0*np.log10(np.abs(freq)/32767.0) # Units of dBFS

plt.plot(freqSteps[0:2000],mag[0:2000])
plt.xlabel("Frequency(MHz)")
plt.grid()
```
